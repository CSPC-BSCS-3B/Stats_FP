#### <font color='Yellow'>**Grouping Job Titles into Categories**</font>

```python
def categorize_job(title):
    title = str(title).lower()
    if 'software' in title or 'developer' in title or 'engineer' in title:
        return 'Technology'
    elif 'data' in title or 'analyst' in title:
        return 'Data & Analytics'
    elif 'marketing' in title or 'media' in title or 'brand' in title:
        return 'Marketing & Comm'
    elif 'nurse' in title or 'therapist' in title or 'practitioner' in title:
        return 'Healthcare'
    elif 'admin' in title or 'assistant' in title or 'coordinator' in title:
        return 'Administration'
    elif 'finance' in title or 'accountant' in title or 'banker' in title:
        return 'Finance'
    elif 'manager' in title or 'project' in title or 'operations' in title:
        return 'Management'
    elif 'support' in title or 'customer' in title or 'service' in title:
        return 'Customer Service'
    else:
        return 'Other'

df['Job_Category'] = df['Job_Title'].apply(categorize_job)
```

#### <font color='Yellow'>**Salary Range vs Country**</font>

```python
# Salary Range vs Country: Heatmap of Frequency (Top 10 Countries incl. Philippines)
# Get top 9 countries by job count, always include Philippines
country_counts = df['Country'].value_counts()
top_countries = list(country_counts.head(9).index)
if 'Philippines' not in top_countries:
    top_countries.append('Philippines')

# Filter data
df_top = df[df['Country'].isin(top_countries)]

# Create a pivot table: countries vs. salary ranges (count of jobs)
pivot = pd.pivot_table(df_top, index='Salary_Range', columns='Country', values='Job_Id', aggfunc='count', fill_value=0)

# Show only the most common salary ranges (top 10 by total count)
top_salary_ranges = pivot.sum(axis=1).sort_values(ascending=False).head(10).index
pivot = pivot.loc[top_salary_ranges]

plt.figure(figsize=(16, 8))
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Job Count by Salary Range and Country (Top 10 Countries incl. Philippines)')
plt.xlabel('Country')
plt.ylabel('Salary Range')
plt.tight_layout()
plt.show()
```

#### <font color='Yellow'>**Job Count by Salary Range and Country**</font>

```python
# Bar Graph: Job Count by Salary Range and Country (Top 10 Countries with Highest Salary Range)
# Convert Salary_Range to numeric midpoint for ranking countries
import re

def extract_mid_salary(s):
    if isinstance(s, str):
        nums = re.findall(r'\$([\d,.]+)[Kk]', s)
        if len(nums) == 2:
            return (float(nums[0].replace(',', '')) + float(nums[1].replace(',', ''))) / 2
        elif len(nums) == 1:
            return float(nums[0].replace(',', ''))
    return None

# Add a numeric midpoint column for salary
mid_salary = df['Salary_Range'].apply(extract_mid_salary)
df_salary = df.copy()
df_salary['Salary_Mid'] = mid_salary

# Compute average salary midpoint per country
country_salary = df_salary.groupby('Country')['Salary_Mid'].mean().sort_values(ascending=False)
top10_salary_countries = country_salary.head(10).index

# Filter data for these countries
salary_df_top = df_salary[df_salary['Country'].isin(top10_salary_countries)]

# Prepare pivot for bar plot (same as before, but for these countries)
pivot = pd.pivot_table(salary_df_top, index='Salary_Range', columns='Country', values='Job_Id', aggfunc='count', fill_value=0)
# Show only the most common salary ranges (top 10 by total count)
top_salary_ranges = pivot.sum(axis=1).sort_values(ascending=False).head(10).index
pivot = pivot.loc[top_salary_ranges]
pivot_long = pivot.reset_index().melt(id_vars='Salary_Range', var_name='Country', value_name='Job_Count')

plt.figure(figsize=(22, 8))
sns.barplot(data=pivot_long, x='Country', y='Job_Count', hue='Salary_Range')
plt.title('Job Count by Salary Range and Country (Top 10 Countries with Highest Avg. Salary Range)\n Which countries offer the highest average salaries, and how are their job postings distributed by salary range?')
plt.xlabel('Country')
plt.ylabel('Job Count')
plt.legend(title='Salary Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
