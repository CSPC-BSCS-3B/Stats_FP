## Data Cleaning

```python
# Renaming Columns
df.rename(columns={ 
    'Job Id': 'Job_Id',
    'Salary Range': 'Salary_Range',
    'Work Type': 'Work_Type',
    'Company Size': 'Company_Size',
    'Job Posting Date': 'Job_Posting_Date',
    'Contact Person': 'Contact_Person',
    'Job Title': 'Job_Title',
    'Job Portal': 'Job_Portal',
    'Job Description': 'Job_Description',
    'Company Profile': 'Company_Profile'
}, inplace=True)
```

```python
df.isnull().sum() # Handling Null Values
```

Only the `Company_Profile` column has null values (5478 rows), which we will drop to retain clean text data for NLP.

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Removing Duplicates
df.drop_duplicates(inplace=True)
```

## Data Visualization

```python
# Work Type Distribution
plt.figure(figsize=(10,5))
sns.countplot(y='Work_Type', data=df, order=df['Work_Type'].value_counts().index)
plt.title("Distribution of Work Types")
plt.xlabel("Count")
plt.ylabel("Work Type")
plt.show()
```

All work types‚ÄîPart-Time, Temporary, Intern, Contract, Full-Time‚Äîappear to have almost equal representation. This is unusual in real-world data, so this might be a balanced sample. 

```python
# Top 10 Countries by Job Listings
top_countries = df['Country'].value_counts().nlargest(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Number of Job Listings")
plt.xlabel("Number of Jobs")
plt.ylabel("Country")
plt.show()
```

Countries like Malta, Venezuela, Somalia, Sierra Leone, Vietnam, and others show up. These aren‚Äôt the typical "tech hub" countries, which suggests that the job data is coming from a global aggregator. 

```python
# Most Common Qualifications
plt.figure(figsize=(10,5))
sns.countplot(y='Qualifications', data=df, order=df['Qualifications'].value_counts().head(10).index)
plt.title("Top 10 Qualifications in Job Postings")
plt.xlabel("Count")
plt.ylabel("Qualification")
plt.show()
```

Top qualifications include BBA, BA, BCA, M.Tech, PhD, MBA, etc.‚Äîwhich shows that both graduate and postgraduate levels are in high demand.

There‚Äôs a strong opportunity for BBA/BCA/MBA students. Great for tailoring career counseling or skill-based upskilling platforms.

```python
# Popular Job Titles
plt.figure(figsize=(10,5))
df['Job_Title'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Job Titles")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.show()
```

UX/UI Designer stands out with a massive lead, followed by titles like Digital Marketing Specialist, Software Engineer, and Network Engineer.

```python
# Company Size Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Company_Size'], bins=30, kde=True)
plt.title("Distribution of Company Size")
plt.xlabel("Company Size")
plt.ylabel("Frequency")
plt.show()
```

This histogram shows a very even distribution across company sizes, which again is uncommon unless data is normalized.
Shows job availability is not skewed towards only startups or MNCs.

## Text Data Exploration

```python
from wordcloud import WordCloud

# WordCloud of Job Titles
text = " ".join(df['Job_Title'].dropna().astype(str))
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(text)
plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Frequent Job Titles")
plt.show()
```

## Group Job Titles into Categories

```python
def categorize_job(title):
    title = str(title).lower()
    if 'software' in title or 'developer' in title or 'engineer' in title:
        return 'Technology'
    elif 'data' in title or 'analyst' in title:
        return 'Data & Analytics'
    elif 'marketing' in title or 'media' in title or 'brand' in title:
        return 'Marketing & Communications'
    elif 'nurse' in title or 'therapist' in title or 'practitioner' in title:
        return 'Healthcare'
    elif 'admin' in title or 'assistant' in title or 'coordinator' in title:
        return 'Administration'
    elif 'finance' in title or 'accountant' in title or 'banker' in title:
        return 'Finance'
    elif 'manager' in title or 'project' in title or 'operations' in title:
        return 'Management'
    elif 'support' in title or 'customer' in title or 'service' in title:
        return 'Customer Service'
    else:
        return 'Other'

df['Job_Category'] = df['Job_Title'].apply(categorize_job)
```

```python
# Frequency Distribution Bar Plot
top_jobs = df['Job_Title'].value_counts().head(20)
plt.figure(figsize=(10, 8))
top_jobs.plot(kind='barh', color='teal')
plt.title('Top 20 Most Frequent Job Titles')
plt.xlabel('Frequency')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Job Categories Distribution
job_cats = df['Job_Category'].value_counts()
plt.figure(figsize=(10, 6))
job_cats.plot(kind='bar', color='coral')
plt.title('Job Categories Distribution')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
